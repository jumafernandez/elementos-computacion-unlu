# 01. Empezando a programar

En esta primera notebook vamos a dar los primeros pasos en Python. El objetivo no es aprender todo el lenguaje de una vez, sino entender cómo trabajar en una notebook, cómo ejecutar código y cómo utilizar variables para representar información.


## 1. ¿Qué es una notebook?

Una notebook es un entorno de trabajo que combina **texto**, **código** y **resultados** en un mismo documento.  
Esto permite explicar una idea y, al mismo tiempo, probarla directamente con Python.

Una notebook está formada por **celdas**. Las más importantes son:

- **Celdas de texto (Markdown):** sirven para escribir explicaciones, títulos, consignas o comentarios.
- **Celdas de código:** sirven para escribir instrucciones en Python y ejecutarlas.

### Idea importante
En una notebook no alcanza con escribir código: hay que **ejecutarlo** para que Python lo procese.


## 2. Ejecutar una celda

Para ejecutar una celda de código se puede usar:

- el botón **Run**
- o la combinación **Shift + Enter**

Probemos con nuestro primer programa.


````python
print("Hola, mundo")
````

La instrucción `print()` se utiliza para **mostrar información en pantalla**.

En este caso, Python muestra el texto `"Hola, mundo"` como salida.


## 3. La notebook recuerda lo que ejecutamos

Cuando ejecutamos una celda, las variables y resultados quedan disponibles para las siguientes celdas.

Por eso, en una notebook importa no solo **qué código escribimos**, sino también **en qué orden lo ejecutamos**.


````python
mensaje = "Estoy empezando a programar"
print(mensaje)
````

Si ahora usamos la variable `mensaje` en otra celda, Python la recordará porque ya fue creada.


````python
print(mensaje)
````

## 4. Variables

Una **variable** es un nombre que usamos para guardar un valor.

Por ejemplo, en la siguiente celda guardamos el número `5` en la variable `x`.


````python
x = 5
print(x)
````

### Atención: el signo `=`

En programación, el signo `=` no representa una igualdad matemática, sino una **asignación**.

Cuando escribimos:

```python
x = 5
```

estamos indicando que la variable `x` pasa a guardar el valor `5`.


## 5. Nombres de variables

Los nombres de las variables deberían ayudar a entender qué información estamos guardando. Por ejemplo:


````python
edad = 35
altura = 1.78
nombre = "Ana"

print(edad)
print(altura)
print(nombre)
````

Usar nombres claros mejora mucho la lectura del programa.


## 6. Tipos de datos básicos

Python puede trabajar con distintos tipos de datos. En esta etapa nos interesan especialmente cuatro:

- **int**: números enteros
- **float**: números con decimales
- **str**: texto
- **bool**: valores lógicos (`True` o `False`)


````python
a = 10
b = 2.5
c = "hola"
d = True

print(a)
print(b)
print(c)
print(d)
````

También podemos preguntarle a Python qué tipo de dato tiene cada variable mediante la función `type()`.


````python
print(type(a))
print(type(b))
print(type(c))
print(type(d))
````

## 7. Operaciones con números

Python permite realizar operaciones aritméticas de forma directa.


````python
x = 8
y = 3

print(x + y)   # suma
print(x - y)   # resta
print(x * y)   # multiplicación
print(x / y)   # división
print(x ** y)  # potencia
````

Esto resulta muy natural para quienes vienen del campo de la matemática, porque muchas expresiones se escriben de manera similar a como las pensamos habitualmente.


## 8. Variables y expresiones

También podemos combinar variables dentro de una expresión y guardar el resultado en una nueva variable.


````python
nota_1 = 7
nota_2 = 9

promedio = (nota_1 + nota_2) / 2
print(promedio)
````

Acá aparece una idea importante: una variable puede guardar no solo un dato escrito “a mano”, sino también el resultado de un cálculo.


## 9. Reasignación de variables

En programación, una variable puede cambiar de valor.


````python
x = 5
print(x)

x = 8
print(x)
````

La variable `x` primero guardaba `5` y luego pasó a guardar `8`.

Esto se llama **reasignación**.


## 10. Un caso especial: `x = x + 1`

Esta expresión suele llamar la atención al principio:


````python
x = 5
x = x + 1
print(x)
````

En matemática, la expresión `x = x + 1` no tendría sentido como igualdad.  
En programación sí tiene sentido porque se interpreta así:

1. tomar el valor actual de `x`
2. sumarle `1`
3. guardar el nuevo resultado en `x`

Es decir, la variable se actualiza.


## 11. Trabajar con texto

Python también permite guardar y mostrar texto.


````python
nombre = "Juan"
saludo = "Hola"

print(nombre)
print(saludo)
````

Podemos mostrar varios elementos juntos con `print()`.


````python
nombre = "Juan"
print("Hola", nombre)
````

## 12. Entrada de datos con `input()`

La función `input()` permite que el programa reciba información escrita por el usuario.


````python
nombre = input("¿Cómo te llamás? ")
print("Hola", nombre)
````

Cuando usamos `input()`, el valor ingresado se interpreta inicialmente como texto.


## 13. Del esquema entrada-proceso-salida al código

Muchos programas sencillos pueden pensarse de esta manera:

- **Entrada:** datos que recibe el programa
- **Proceso:** operaciones que realiza
- **Salida:** resultado que muestra

Veamos un ejemplo simple.


````python
numero_1 = 6
numero_2 = 4

suma = numero_1 + numero_2

print("El resultado es:", suma)
````

En este ejemplo:

- la **entrada** son los valores de `numero_1` y `numero_2`
- el **proceso** es la suma
- la **salida** es el valor mostrado con `print()`


## 14. Ejemplo aplicado: calcular el área de un rectángulo


````python
base = 5
altura = 3

area = base * altura
print("El área del rectángulo es:", area)
````
